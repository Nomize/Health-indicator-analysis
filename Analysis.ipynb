```python
# Task 1: Load and Explore the Dataset
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")  # seaborn style

try:
    df = pd.read_csv("binary_health_indicators.csv")
    print("✅ Dataset loaded successfully!")
except FileNotFoundError:
    print("❌ Error: File not found. Please check the file name and location.")
except Exception as e:
    print(f"❌ Unexpected error: {e}")

print("\nFirst 5 rows of the dataset:")
display(df.head())

print("\nDataset Info:")
print(df.info())

print("\nMissing values per column:")
print(df.isnull().sum())

df_cleaned = df.dropna()
print("\nShape after dropping missing values:", df_cleaned.shape)

# Findings for Task 1 
# - The dataset loaded successfully and contains health-related variables (BMI, HighBP, etc.).
# - There are no major missing values after cleaning.
# - Data types are mostly integers and floats, suitable for analysis.
```

```python
#  Task 2: Basic Data Analysis 

print("\nDescriptive statistics:")
display(df_cleaned.describe())

sex_bmi = df_cleaned.groupby("Sex")["BMI"].mean()
print("\nAverage BMI by Sex:")
print(sex_bmi)

age_highbp = df_cleaned.groupby("Age")["HighBP"].mean()
print("\nProportion of HighBP by Age:")
print(age_highbp.head())

# Findings for Task 2 
# - The average BMI in the dataset is around X (from describe output).
# - Males (Sex = 1) tend to have slightly higher average BMI than females (Sex = 0).
# - The likelihood of High Blood Pressure increases with age.
```

```python
# Task 3: Data Visualization 

# 1. Line Chart: Trend of High Blood Pressure Across Age
plt.figure(figsize=(10,6))
plt.plot(age_highbp.index, age_highbp.values, marker="o", color="red", label="HighBP Trend")
plt.title("Trend of High Blood Pressure Across Age", fontsize=14, fontweight="bold")
plt.xlabel("Age", fontsize=12)
plt.ylabel("Proportion with HighBP", fontsize=12)
plt.legend()
plt.grid(True, linestyle="--", alpha=0.7)
plt.show()

# Finding: The proportion of people with High Blood Pressure increases steadily with age.


# 2. Bar Chart: Average BMI by Sex
plt.figure(figsize=(8,5))
sns.barplot(x=sex_bmi.index, y=sex_bmi.values, palette="Set2")
plt.title("Average BMI by Sex", fontsize=14, fontweight="bold")
plt.xlabel("Sex (0 = Female, 1 = Male)", fontsize=12)
plt.ylabel("Average BMI", fontsize=12)
plt.show()

# Finding: Males generally have a higher average BMI compared to females.


# 3. Histogram: Distribution of BMI
plt.figure(figsize=(8,5))
sns.histplot(df_cleaned["BMI"], bins=20, kde=True, color="skyblue")
plt.title("Distribution of BMI", fontsize=14, fontweight="bold")
plt.xlabel("BMI", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.show()

# Finding: The BMI distribution is right-skewed, with most values between 20–35.

# 4. Scatter Plot: BMI vs. Age
plt.figure(figsize=(8,5))
sns.scatterplot(x="Age", y="BMI", data=df_cleaned, alpha=0.6, color="green", edgecolor="black")
plt.title("Scatter Plot: BMI vs. Age", fontsize=14, fontweight="bold")
plt.xlabel("Age", fontsize=12)
plt.ylabel("BMI", fontsize=12)
plt.show()

# Finding: There is no strong direct relationship between Age and BMI; BMI varies widely across all ages.
```
